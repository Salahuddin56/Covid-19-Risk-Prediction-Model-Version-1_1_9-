```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from pandas.plotting import parallel_coordinates
```

```python
# read excel file and create pandas dataframe
df = pd.read_excel('Final-Data-COVID-PATIENT-STATE-2020-2022-dirty-data2.xlsx')
```

```python
# --------- data preprocessing (e.g. data cleaning) -------------

# ... (all the previous data preprocessing code)

# Convert the 'Age' column to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Drop rows with missing age values
df.dropna(subset=['Age'], inplace=True)

# Create a new 'Age Range' column based on the specified age ranges
age_ranges = {
    (1, 12): 'Children',
    (13, 17): 'Adolescents',
    (18, 64): 'Adults',
    (65, 100): 'Elders'
}

def categorize_age(age):
    for range, label in age_ranges.items():
        if range[0] <= age <= range[1]:
            return label

df['Age Range'] = df['Age'].apply(categorize_age)

# Remove the 'Age' column since we now have the 'Age Range' column
df.drop('Age', axis=1, inplace=True)

# Encode the 'Age Range' column using one-hot encoding
df = pd.get_dummies(df, columns=['Age Range'])

# Feature matrix
X = df[['S1: Fever', 'S2: Cough', 'S3: Joint Pain', 'Test2:RT PCR', 'Test3: Chest X Ray', 'Test4: CT Scan ', 'Status']]

# Convert all columns to numeric
X = X.apply(pd.to_numeric, errors='coerce')

# Impute missing values in numeric columns with the median value
imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Convert 'Status' column to numeric
X['Status'] = pd.to_numeric(df['Status'], errors='coerce')

# Target matrix
y = df['Risk Factor']
```

```python


# Model Training and Cross-validation

models = {
    "SVM": svm.SVC(),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier()
}

scores = {}

# Perform cross-validation for each model
for model_name, model in models.items():
    cv_scores = cross_val_score(model, X, y, cv=5)  # Perform 5-fold cross-validation
    avg_score = np.mean(cv_scores).round(4)
    scores[model_name] = avg_score

print("----------------------------------------")
for model_name, score in scores.items():
    print(f"Average Cross-validated Accuracy of {model_name}: {score*100:.2f}%")
print("----------------------------------------")

# Plot the cross-validated accuracy scores of different models
data = {
    "Models": list(scores.keys()),
    "Cross-validated Accuracy(%)": [score * 100 for score in scores.values()]
}
cv_score_df = pd.DataFrame(data)
print(cv_score_df)

sns.barplot(x="Models", y="Cross-validated Accuracy(%)", data=cv_score_df, palette="plasma")
plt.title("Cross-validated Accuracy of Five models")
plt.show()
```
